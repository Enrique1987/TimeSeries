### 5.1 Moving Average Smoothing

Smoothing is a technique applied to TS to remove variation between steps. The hope of smoothing is to remove noise and better expose the signal of underlying causal process. Moving averages are a simple and common type of smoothing used in TSA and TSF. Moving average requires that we specify Windows-size.

There are two main types of moving average that are used.

   - Centered
   - Trailing Moving Average.
   

#### Centered Moving Average

The value at time(t)

center_ma(t) = mean(obs(t − 1), obs(t), obs(t + 1))

#### Trailing Moving Average

The value at time(t)

trail ma(t) = mean(obs(t − 2), obs(t − 1), obs(t))
