### 5.1 Moving Average Smoothing

Smoothing is a technique applied to TS to remove variation between steps. The hope of smoothing is to remove noise and better expose the signal of underlying causal process.


#### Airline Passengers Dataset

This dataset describes the total number of airline passengers over time.

```python
from pandas import read_csv
from datetime import datetime
from matplotlib import pyplot
pyplot.rcParams['figure.figsize'] = [13, 7]
import os

data = os.environ.get('data') # avoid personal information in the notebook
data_path = data + '\\01_TimeSeries\\2-airline-passengers.csv'
series = read_csv(data_path, header=0, index_col=0, parse_dates=True,squeeze=True)
series.head(n=7)
```

```python
pyplot.figure(1)
# line plot
pyplot.subplot(211)
pyplot.plot(series)
# histogram
pyplot.subplot(212)
pyplot.hist(series)
pyplot.show()
```

#### Square Root Transform

TS with a quadratic growth can be madre linear by taking the square root. Lets demostrate it with a example

   - Series numbes 1 to 99 squared,
   - The line plot will show a cuadratic growth.

```python
from matplotlib import pyplot
series = [i**2 for i in range(1,100)]
pyplot.figure(1)
```

```python
# line plot
pyplot.subplot(211)
pyplot.plot(series)
```

```python
pyplot.subplot(212)
pyplot.hist(series)
pyplot.show()
```

If you see a structure like this in your own time series, you may have a quadratic growth trend. This can be remove by taking the inverse operation of the squaring preocedure which is the square root.

```python
from numpy import sqrt
series = [i**2 for i in range(1,100)]
# sqrt transform
transform = sqrt(series)
pyplot.figure(1)
# line plot
pyplot.subplot(211)
pyplot.plot(transform)
# histogram
pyplot.subplot(212)
pyplot.hist(transform)
pyplot.show()
```
