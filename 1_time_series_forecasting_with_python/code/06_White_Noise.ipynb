### 6.1 What is White Noise and why does it Matter ? 

A TS may be white noise. TS is White Noise if the varibles are independent and identically distributed with a mean of zero.

This means that all varibles have the same variance(sigma2) and each value has a zero correlation with all others values in the series. If the variable in teh series are drawn from a Gaussian distribution, the sereis is calles Gaussian white noise.

- **Predictability:** If your Ts is whit noise is random you can not make predictions reasonably.
- **Model Diagnostics:** The series o erros from a TS forecast model should ideally be white nose.

Model diagnostics is an imporant area of TSF. Time series data are expected to contain some white noise component on top of the signal.

### 6.2 Is your TS White Noise ?

Your TS is not white nose if any of the following conditiosn are true:

- Does your seires have a non-zero mean ?

- Does the variance change over time ?

- Does values corelate with lag values ?

Some tools that you can use to check if your TS is whit noise are:

- **Create a line plot:** Check for gross features like a changin mean, variance, or obviouss relationsip between lagge variables.
- **Calculate summary statistics:** Check the mean and variance of the whole series against the mean and variance of meaningful contignuous blocks of variables.
- **Create an autocorrelation plot:** Check for gross correlation between lagged variables.

In dis section we will create a Gaussian white noise. We can create variables using the gauss() function from the random module

```python
from random import gauss
from random import seed
from pandas import Series
from pandas.plotting import autocorrelation_plot
from matplotlib import pyplot
# seed random number generator
seed(1)
# create white noise series
series = [gauss(0.0, 1.0) for i in range(1000)]
series = Series(series)
```

```python
# summary stats
print(series.describe())
```

The standard `normal distribution`, also called the z-distribution, is a special normal distribution where the **mean is 0 and the standard deviation is 1**.

```python
# line plot
series.plot()
pyplot.show()
```

```python
# to appreciate the Gaussian distribution we need a histogram
# line plot
series.hist()
pyplot.show()
```

```python
# autocorrelation
pyplot.rcParams['figure.figsize'] = [13, 7]
autocorrelation_plot(series)
pyplot.show()
```

The correlogram does not show any obvious autocorrelation pattern. There are some spikes above the 95% and 99% confidence level, but thes are statical fluke.
